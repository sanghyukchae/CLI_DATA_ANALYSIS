```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df= pd.read_csv("sst_anomaly_g_new.csv", encoding="cp949")

df.drop(["Unnamed: 2"],axis=1,inplace=True)


df["DATE"] = df["연도"].astype(str)
df["SST anomaly"]= df["전지구 연평균 해수면온도 편차"]

df.drop(["연도"],axis=1,inplace=True)
df.drop(["전지구 연평균 해수면온도 편차"],axis=1,inplace=True)




cond1= df["DATE"].str.contains("1")
aa=df.loc[cond1] 

cond2= df["DATE"].str.contains("2")

aa=df.loc[cond1|cond2] 


plt.figure(figsize=(12,5))



at1 = np.array(aa["SST anomaly"])
at2 = np.array(aa["DATE"])

plt.plot(at2,at1,label="℃", c="r")
plt.xlabel("Year")


plt.xticks(["1891","2020","1954","1923","1987"])

plt.legend(loc="lower right")
plt.title("SST Anomaly (Global)")

plt.show()


#Global Sea Surface Temperature Anomaly

#Source: 
#Korea Meteorological Adminstration
#Japan Meteorological Agency
#http://www.climate.go.kr/home/09_monitoring/info/data_download
#https://www.data.jma.go.jp/gmd/kaiyou/data/shindan/a_1/glb_warm/glb_warm.html
```
