```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df= pd.read_csv("sst_area_k_new.csv",encoding="cp949")

df.drop(["Unnamed: 4"],axis=1,inplace=True)


df["YEAR"]= df[""].astype(str)
df["The East Sea"]=df[""]
df["The Southern Sea"]=df[""]
df["The Western Sea"]=df[""]


df.drop([""],axis=1,inplace=True)
df.drop([""],axis=1,inplace=True)
df.drop([""],axis=1,inplace=True)
df.drop([""],axis=1,inplace=True)


cond1= df["YEAR"].str.contains("1")
cond2= df["YEAR"].str.contains("2")
aa=df.loc[cond1|cond2] 


plt.figure(figsize=(12,5))

at1 = np.array(aa["The East Sea"])
at2 = np.array(aa["YEAR"])

at3 = np.array(aa["The Southern Sea"])
at2 = np.array(aa["YEAR"])

at4 = np.array(aa["The Western Sea"])
at2 = np.array(aa["YEAR"])


plt.plot(at2,at1,label="The East Sea", c="r")
plt.plot(at2,at3,label="The Southern Sea", c="b")
plt.plot(at2,at4,label="The Western Sea", c="y")


plt.xlabel("Year")
plt.ylabel("")

plt.xticks(["1968","2020","1994","1981","2007"])

plt.legend(loc="lower right")
plt.title("SST")

plt.show()



#Source:
#Korea Meteorological Adminstration -> http://www.climate.go.kr/home/09_monitoring/info/data_download
#Korea Oceaonographic Data Center -> https://www.nifs.go.kr/kodc/soo_list.kodc

```
